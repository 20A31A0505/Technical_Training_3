```python
#FACTORIAL OF A NUMBER USING RECURSION
def fact(n):
    if n==0 or n==1:
        return 1
    else:
       return n*fact(n-1)
n=int(input("Enter the input:"))
print(fact(n))
```

```python
#SUM OF FIRST 5 NATURAL NUMBERS USING RECURSION
def sum(n):
    if n==0 or n==1:
        return 1
    else:
       return n+sum(n-1)
n=int(input("Enter the input:"))
print(sum(n))
```

```python
#FIBONOCCI SERIES USING RECURSION
def fib(n):
  if n<0:
    print("Incorrect")
  elif n==0:
    return 0
  elif n==1 or n==2:
    return 1
  else:
    return fib(n-1)+fib(n-2)
n=int(input("Enter the length of the fibbinocci series :"))
print(fib(n))
```

```python
#Find the power of a number using recursion
def pow(a,b):
  if b==0:
    return 1
  else:
    return (a*pow(a,b-1))
a=int(input("Enter the base value:"))
b=int(input("Enter the exp value: "))
print(pow(a,b))
```

```python
#ARMSTRONG NUMBER
def count(n):
  c=0
  while n>0:
    n1=n%10
    n//=10
    c+=1
  return c
n=int(input("Enter a number:"))
sum=0
c=count(n)
temp=n
while temp>0:
  digit=temp%10
  sum+=digit**c
  temp//=10
if n==sum:
  print(n," is an armstrong number")
else:
  print(n ,"is not a armstrong number")
```

```python
#ARMSTRONG NUMBER IN A INTERVAL
def count(n):
  c=0
  while n>0:
    n1=n%10
    n//=10
    c+=1
  return c
def armstrong(n):
  n2=n
  t=0
  c=count(n)
  while n>0:
    n1=n%10
    t+=n1**c
    n//=10
  if n2==t:
    return t
n,n1=map(int,input("Enter the range :").split(" "))
for i in range(n,n1+1):
  if armstrong(i)!=None:
    print(armstrong(i),end=" ")
```

```python
#MEMOIZATION OF FIBONACII SERIES
#TOP DOWN APPROACH
def fib(n,dp):
  if n==0:
    return 0
  elif n==1:
    return 1
  elif dp[n]!=-1:
    return dp[n]
  else:
    dp[n]=fib(n-1,dp)+fib(n-2,dp)
    return dp[n]
n=int(input())
dp=[-1]*n
print(fib(5,dp))
print(dp)
```
